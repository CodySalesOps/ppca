```python
os.sys.path.append('../src/')
import ppca
```

```python
def plot_data(x):
    fig, ax = plt.subplots()
    col_map = {0:'red', 1:'green', 2:'blue'}
    colors = [col_map[i] for i in y_class]
    ax.scatter(x[0, :], x[1, :], color=colors)
```

```python
iris = ds.load_iris()
y = np.transpose(iris.data)
y_class = iris.target
```

# Maximum likelikhood

```python
pca_ml = ppca.PPCA()
x = pca_ml.fit_transform(y)
plot_data(x)
```
